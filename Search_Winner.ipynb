```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
sl = glob.glob('./stock/*.csv')
lst = pd.read_csv('all_list.csv')
slist = list(lst['Kode']+".JK")
```

```python
#variable kenaikan. 1.2 = 20%
uprise=1.1

sql = []
for i in sl:
    df = pd.read_csv(i)
    
    try:
        dlast = df.iloc[-1]['Close']
        dprev = df.iloc[-2]['Close']

        if (dlast>dprev*uprise):
            print(i)
            sql.append(i)
    except:
        pass
```

```python
fig, ax = plt.subplots(nrows=len(sql),figsize=(10,30))

for n,i in enumerate(sql):
    stk = pd.read_csv(i)
    ax[n].plot(stk['Close'])
    ax[n].set_title(i)
    fig.tight_layout()
```
